
Overview of OpenWeatherMap API
Current Weather Data: Provides current weather information for any location, including over 200,000 cities.
5 Day / 3 Hour Forecast: Provides weather forecast data for the next 5 days with 3-hour intervals.
16 Day / Daily Forecast: Provides daily weather forecast data for up to 16 days.
Historical Weather Data: Provides historical weather data for a specific location.
Weather Alerts: Provides severe weather alerts.
Key API Endpoints

# Current Weather Data:
## Endpoint: http://api.openweathermap.org/data/2.5/weather
Parameters:
q: City name (e.g., q=London)
id: City ID
lat, lon: Latitude and Longitude
zip: Zip code
appid: Your API key
units: Units of measurement (standard, metric, imperial)
lang: Language for the response
Documentation: Current Weather Data

# 5 Day / 3 Hour Forecast:
## Endpoint: http://api.openweathermap.org/data/2.5/forecast
Parameters: Similar to the Current Weather Data
Documentation: 5 Day / 3 Hour Forecast

# 16 Day / Daily Forecast:
## Endpoint: http://api.openweathermap.org/data/2.5/forecast/daily
Parameters: Similar to the Current Weather Data
Documentation: 16 Day / Daily Forecast

# Historical Weather Data:
## Endpoint: http://api.openweathermap.org/data/2.5/onecall/timemachine
Parameters: Latitude and Longitude, Time, API key
Documentation: Historical Weather Data

# Weather Alerts:
## Endpoint: Integrated in One Call API
Documentation: Weather Alerts

# I Current weather data

```python
import requests
import pandas as pd

api_key = "554897c795df8884a3af6b30df9c8e98"

def get_weather(city, api_key): 
    base_url = "http://api.openweathermap.org/data/2.5/weather"
    params = {
        'q': city,
        'appid': api_key,
        'units': 'metric'  # to get the data in degrees Celsius
    }
    
    response = requests.get(base_url, params=params)
    if response.status_code == 200: 
        data = response.json()
        weather = {
            'city': data['name'],
            'temperature': data['main']['temp'],
            'description': data['weather'][0]['description'],
            'humidity': data['main']['humidity'],
            'pressure': data['main']['pressure'],
            'wind_speed': data['wind']['speed']
        }
        return weather
    else: 
        print(f"Failed to get weather data for {city}. HTTP Status code: {response.status_code}")
        return None

def collect_weather_data(cities, api_key):
    weather_data = []
    for city in cities:
        weather = get_weather(city, api_key)
        if weather:
            weather_data.append(weather)
    return weather_data

# Define your cities
cities = ['Paris', 'Jamshedpur', 'Kolkata']

# Collect weather data
weather_data = collect_weather_data(cities, api_key)

# Convert the weather data to a DataFrame
weather_df = pd.DataFrame(weather_data)

# Display the DataFrame
# print(weather_df)
weather_df
```

```python
import pytz

def get_coordinates(city, api_key):
    base_url = "http://api.openweathermap.org/geo/1.0/direct"
    params = {
        'q': city,
        'appid': api_key,
        'limit': 1  # To ensure we get only the most relevant result
    }
    
    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        data = response.json()
        if data:
            return data[0]['lat'], data[0]['lon']
    print(f"Failed to get coordinates for {city}. HTTP Status code: {response.status_code}")
    return None, None

def get_air_pollution_forecast(lat, lon, api_key):
    base_url = "http://api.openweathermap.org/data/2.5/air_pollution/forecast"
    params = {
        'lat': lat,
        'lon': lon,
        'appid': api_key
    }
    
    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        return response.json()
    print(f"Failed to get air pollution data for coordinates ({lat}, {lon}). HTTP Status code: {response.status_code}")
    return None

def process_air_pollution_data(city, data):
    forecasts = []
    for entry in data['list']:
        utc_time = datetime.utcfromtimestamp(entry['dt'])
        local_time = utc_time.replace(tzinfo=pytz.utc).astimezone(pytz.timezone('Europe/Paris'))  # Adjust time zone accordingly
        local_time_str = local_time.strftime('%Y-%m-%d %H:%M:%S %Z%z')
        
        forecast = {
            'city': city,
            'aqi': entry['main']['aqi'],
            'co': entry['components']['co'],
            'no': entry['components']['no'],
            'no2': entry['components']['no2'],
            'o3': entry['components']['o3'],
            'so2': entry['components']['so2'],
            'pm2_5': entry['components']['pm2_5'],
            'pm10': entry['components']['pm10'],
            'nh3': entry['components']['nh3'],
            'local_time': local_time_str
        }
        forecasts.append(forecast)
    return forecasts

def collect_air_pollution_data(cities, api_key):
    all_forecasts = []
    for city in cities:
        lat, lon = get_coordinates(city, api_key)
        if lat is not None and lon is not None:
            pollution_data = get_air_pollution_forecast(lat, lon, api_key)
            if pollution_data:
                forecasts = process_air_pollution_data(city, pollution_data)
                all_forecasts.extend(forecasts)
    return all_forecasts

# Define your cities
cities = ['Paris', 'Jamshedpur', 'Kolkata']

# Replace 'your_api_key' with your own API key obtained from OpenWeatherMap
api_key = '554897c795df8884a3af6b30df9c8e98'

# Collect air pollution forecast data
air_pollution_data = collect_air_pollution_data(cities, api_key)

# Convert the air pollution data to a DataFrame
if air_pollution_data:
    air_pollution_df = pd.DataFrame(air_pollution_data)
    # print(air_pollution_df)
else:
    print("No air pollution data available.")
air_pollution_df
```
